# TRANSFORMACIÓN DE DATOS (Feature Engineering)
A partir de una varables, creamos otros datos.

Tabajamos con la base de datos bikes.csv

La información de los datos es la siguiente:

- datetime: hora por hora + marca de tiempo

- season:
    - 1 = primavera
    - 2 = verano
    - 3 = caída
    - 4 = invierno

- holiday: si el día se considera un día festivo

- workingday: si el día no es un fin de semana ni un feriado.

- clima:

    - 1: Despejado, pocas nubes, medio nublado, parcialmente nublado
    - 2: Niebla + Nublado, Niebla + parcialmente nublado, Niebla + Pocas nubes, Niebla
    - 3: Nieve ligera, Lluvia ligera + Tormenta eléctrica + Nubes dispersas, Lluvia ligera + Nubes dispersas
    - 4: Lluvia intensa + Paletas de hielo + Tormenta + Niebla, Nieve + Niebla


- temp: temperatura en grados Celsius

- atemp: "sensación térmica" la temperatura en grados Celsius

- humidity: humedad relativa

- windspeed: velocidad del viento

- casual: número de alquileres de usuarios no registrados iniciados

- registered: número de alquileres de usuarios registrados iniciados

- count: número de alquileres totales

```python
def asignaEstacion(variable):
  if  variable == 1:
    return "primavera"
  elif variable == 2:
    return "verano"
  elif variable==3:
    return "otoño"
  elif variable ==4:
    return "invierno"
  else:
    return None
```

Tipos de datos (Feature Engineering)

- Numéricos: Aquí estan los float, int
- Temporales: Fechas, id correlativos, etc
- Categóricos:
  - Identificador: DNI,#telefono,placa de carro,#pasaporte,codigo de alumno.
    El valor suele asociarse directamente a un solo ente o fila.
  - Ordinales: Si tienen un orden intrínseco.
      Ejemplo:
        - Primaria, secundaria, universidad, maestría, doctorado
        - triste, serio, feliz
        - a,b,c,d,e,f
        - Lunes, Martes, Miércoles, Jueves, Viernes, Sábado, Domingo
        - Enero, Febrero, Marzo, Abril,

  - Nominales: No tienen un orden intrínseco. No suele identificar a las filas o entidad a analizar. Se suele hacer codificación one-hot-encoding.
        - Rojo, Verde, Azul, Marrón,etc
        - Manzana, Pera, Banana, etc
        - turno mañana, turno noche,etc
        - telefono, laptop, tablet, pc.
        - raza de animales
        - especies de plantas.
        - etc
  - Dicotomicas: tienen 2 valores. Se suele codificar como 1 columna de 0/1.
        - Hombre/Mujer
        - Dia/Noche
        - Moroso/No moroso
        - Venta/No venta
        - Encendido/Apagado
        - Contacto/No contacto (llamada telefónica)
        - Disponible/No disponible

- Textos: Frases u oraciones que expresan un significado a analizar:
        - Reviews de películas :
            - La película X estuvo aburrida y muy lenta, no se entendio el final.
            - Estuvo buenísima, tuvo un final interesante.
            - Duro 2h, es en promedio superior a la mayoría de películas en duración.

        - Tweets:
          - Puedes analizar tópicos de interés.
        
        - Libros o información en párrafos:
          - Hacer resumenes o contestar preguntas en base a un texto (libro,parrafo, etc)

- Imágenes/Video:
        - satelitales: Se suele analizar para predicción de clima.
        - planos geográficos
        - mapas
        - imágenes de videovigilancia

- Audio:
        - Llamadas en call center: se usa para identificar emoción del cliente. detección de palabras clave, etc
        - transcripción de audio a texto, etc

- Grafos:
        - Relaciones de amistad en fb
        - Linkedin
        - Análisis de seguidores en twiter
        - Networking profesional en una empresa por medio de correos.

```python
# library
import seaborn as sns
import pandas as pd
import numpy as np

corr_matrix=df.corr()
# Can be great to plot only a half matrix
# Generate a mask for the upper triangle
mask = np.zeros_like(corr_matrix)
mask[np.triu_indices_from(mask)] = True

# Draw the heatmap with the mask
sns.heatmap(corr_matrix, mask=mask, square=True)
```

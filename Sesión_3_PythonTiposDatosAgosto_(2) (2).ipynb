## Bucles condicionales (while)
* while condición:
*    bloque código

Repite la ejecución del bloque de código mientras la expresión lógica condición sea cierta.

Se puede interrumpir en cualquier momento la ejecución del bloque de código con la instrucción **break**.

 El bloque de código debe estar **indentado por un tab.**

````python
while 1==1:
  print('Hola esto es una prueba')
  print('---')
````

````python
while 1!=1:
  print('Hola esto es una prueba')
  print('---')
````

````python
# Pregunta al usuario por un número hasta que introduce 0.
# != : Signfica diferente
# Condición de aprobar un curso en la universidad sea 13/20

desaprobado=True

while desaprobado==True:
  nota = float(input('Introduce tu nota del curso: '))
  desaprobado = ( nota<13 )
  if desaprobado == True:
    print("Has desaprobado, debes repetir el curso")

print("Has aprobado")
````

````python
estado="no cursado"

while (estado=="no cursado") | (estado=="desaprobado"):
  print("Llevar el curso")
  nota=float(input("La nota es: "))

  if nota<13:
    estado="desaprobado"
  else:
    estado="aprobado"

print("Saliste del bucle")
````

# 4. Bucles iterativos (for)
* for i in secuencia:
*    bloque código

**Repite la ejecución** del bloque de código para cada elemento de la secuencia secuencia, asignado dicho elemento a i en cada repetición.

Se puede interrumpir en cualquier momento la ejecución del bloque de código con la instrucción break o saltar la ejecución para un determinado elemento de la secuencia con la instrucción continue.

## El bloque de código debe estar indentado por 2 espacios o tab.

Se utiliza fundamentalmente para recorrer colecciones de objetos como cadenas, listas, tuplas o diccionarios.

A menudo se usan con la instrucción range:

* range(fin) : Genera una secuencia de números enteros desde 0 hasta fin-1.
* range(inicio, fin, salto) : Genera una secuencia de números enteros desde inicio hasta fin-1 con un incremento de salto.

````python
cualquier_iterable=["asds","eprdf","kswo","qeysd"]
for cualquier_cosa in cualquier_iterable:
  print(cualquier_cosa)
````

````python
palabra = 'Python'

for letra in palabra:
  print(letra)
````

## Ejercicio 15 - BUCLES

Escribir un programa que pida al usuario un número entero positivo y muestre por pantalla todos los números impares desde 1 hasta ese número separados por comas.

````python
n=int(input("Ingrese un número entero positivo:"))
cadena=""
for i in range(1,n,2):
  cadena=cadena+str(i)+","

cadena=cadena[:-1]
print(cadena)
````

## EJERCICIO FINAL

Una empresa desarrolladora de Software ha puesto a la venta licencias de su programa de edición de videos "Video Edit 2.0" a los siguientes costos:


|    Licencia    | Costos |
| :---------: | :--------: |
| Cobre |    $510     |
|  Bronze  |    1500     |
|  Silver  |    3100     |
|  Gold  |    4500     |


Diseñe un programa que permita ingresar por cada venta el tipo de licencia y la cantidad de licencias, y muestre luego cada venta:

* El importe a pagar por la venta efectuada.
* El importe total recaudado por cada tipo de licencia.
* La cantidad de licencias vendidas por cada tipo de licencia.
* La cantidad de ventas por cada tipo de licencia.

````python
print("Ventas")
cantidad_cobre=int(input("ingrese la cantidad de licencias vendidas de Cobre: "))
cantidad_bronze=int(input("ingrese la cantidad de licencias vendidas de Bronze: "))
cantidad_silver=int(input("ingrese la cantidad de licencias vendidas de Silver: "))
cantidad_gold=int(input("ingrese la cantidad de licencias vendidas de Gold: "))

print("el importe a pagar por la venta efectuada es",cantidad_cobre*510+cantidad_bronze*1500+cantidad_silver*3100+cantidad_gold*4500)

print("el importe recaudado por licencias Cobre es",cantidad_cobre*510)
print("el importe recaudado por licencias Bronze es",cantidad_bronze*1500)
print("el importe recaudado por licencias Silver es",cantidad_silver*3100)
print("el importe recaudado por licencias Gold es",cantidad_gold*4500)

print("La cantidad de licencias Cobre vendidas es",cantidad_cobre)
print("La cantidad de licencias Bronze vendidas es",cantidad_bronze)
print("La cantidad de licencias Silver vendidas es",cantidad_silver)
print("La cantidad de licencias Gold vendidas es",cantidad_gold)
````
